```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("data.csv")
```

```python
X = data.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = data.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
```

```python
model=LinearRegression()
```

```python
XTrain,XTest,YTrain,YTest=train_test_split(X,Y,test_size=(1/3),random_state=4)
```

```python
print(XTrain.shape)
print(YTrain.shape)
```

```python
print(XTest.shape)
print(YTest.shape)
```

```python
model.fit(XTrain,YTrain)
```

```python
prediction=model.predict(XTest)
```

```python
prediction
```

```python
model.coef_
```

```python
error=np.mean(np.square(prediction-YTest))
print("Mean square error is : ",end=" ")
print(error)
```

```python
model.score(XTest,YTest)
```

```python
plt.scatter(X,Y)
plt.plot(XTest,prediction,color='red',label='predicted values')
plt.xlabel("Test set")
plt.ylabel("Prediction")
plt.title("LINEAR REGRESSION WITH 1 VARIABLE")
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(X,Y,color='r')
plt.title('100% accuracy graph')
plt.show()
```
